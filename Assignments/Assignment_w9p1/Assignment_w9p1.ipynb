```python
import pandas as pd
import numpy as np

import pandas_profiling

from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
```

```python
x = pd.read_csv('../input/2022-ml-w9p1/x_train.csv',encoding='CP949')
y = pd.read_csv('../input/2022-ml-w9p1/y_train.csv',encoding='CP949')
test = pd.read_csv('../input/2022-ml-w9p1/x_test.csv',encoding='CP949')
```

```python
sc = StandardScaler()
x = sc.fit_transform(x)
test = sc.transform(test)
```

```python
model = SVR(kernel='precomputed')
model.fit(x, y['교통량'])
res = model.predict(test)
res = pd.DataFrame(res, columns=['predict'])
res.to_csv('submit.csv', index_label='id')
```
