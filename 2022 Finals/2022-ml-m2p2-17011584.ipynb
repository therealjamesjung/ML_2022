```python
# 필요한 패키지 import
import pandas as pd
import numpy as np

# 랜덤시드 고정
import random
import os
seed = 42
random.seed(seed)
np.random.seed(seed)
os.environ['PYTHONHASHSEED'] = str(seed)
```

```python
# 데이터 불러오기
x = pd.read_csv('../input/2022-ml-finalexam-p2/NetworkTrain_x.csv')
y = pd.read_csv('../input/2022-ml-finalexam-p2/NetworkTrain_y.csv')
test = pd.read_csv('../input/2022-ml-finalexam-p2/NetworkTest_x.csv', header=None)
```

```python
# column 이름 초기화
x = x.T.reset_index(drop=True).T
test = test.T.reset_index(drop=True).T
```

```python
# 라벨 데이터 실수화
y.loc[y[y['normal'] == 'normal'].index, 'normal'] = 0
y.loc[y[y['normal'] != 0].index, 'normal'] = 1

y = y.astype('int64')
```

```python
# 문자열 데이터 실수화
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
def label_encode(train, test, columns):
    for col in columns:
        labels = pd.concat([train[col], test[col]])
        le.fit(labels)
        train[col] = le.transform(train[col])
        test[col] = le.transform(test[col])
        
label_encode(x, test, [1, 2, 3])
```

```python
# 모델 학습 및 결과 도출
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

rf = RandomForestClassifier(random_state=1, class_weight='balanced')
model = AdaBoostClassifier(rf, random_state=1)
model.fit(x, y)
res = model.predict(test)
submit = pd.read_csv('../input/2022-ml-finalexam-p2/submission_sample.csv')
submit.Category = res
submit.to_csv('submit.csv', index=False)
```
