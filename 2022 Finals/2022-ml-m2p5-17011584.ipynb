```python
# 필요한 패키지 import
import pandas as pd
import numpy as np

# 랜덤시드 고정
import random
import os
seed = 42
random.seed(seed)
np.random.seed(seed)
os.environ['PYTHONHASHSEED'] = str(seed)
```

```python
# 데이터 불러오기
train = pd.read_csv('../input/2022-ml-finalexam-p5/train.csv')
test = pd.read_csv('../input/2022-ml-finalexam-p5/test.csv')

# 라벨 분리 및 필요없는 column drop
y = train['Survived']
train = train.drop(['PassengerId', 'Name', 'Ticket', 'Fare', 'Cabin', 'Parch', 'SibSp', 'Survived'], axis=1)
test = test.drop(['PassengerId', 'Name', 'Ticket', 'Fare', 'Cabin', 'Parch', 'SibSp'], axis=1)
```

```python
# 결측 데이터 채워주기
train.Embarked = train.Embarked.fillna('S')
```

```python
# Imputer 활용 결측데이터 평균값으로 채워주기
from sklearn.impute import SimpleImputer

imp = SimpleImputer()

train.Age = pd.DataFrame(imp.fit_transform(np.array(train.Age).reshape(-1, 1)))
test.Age = pd.DataFrame(imp.fit_transform(np.array(test.Age).reshape(-1, 1)))
```

```python
# 문자열 데이터 실수화
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
def label_encode(train, test, columns):
    for col in columns:
        labels = pd.concat([train[col], test[col]])
        le.fit(labels)
        train[col] = le.transform(train[col])
        test[col] = le.transform(test[col])

label_encode(train, test, ['Embarked', 'Sex'])
```

```python
# 모델 학습 및 결과 도충
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=1, class_weight='balanced')
model.fit(train, y)
res = model.predict(test)
submit = pd.read_csv('../input/2022-ml-finalexam-p5/sample_submit.csv')
submit['Survived'] = res
submit.to_csv('submit.csv', index=False)
```
