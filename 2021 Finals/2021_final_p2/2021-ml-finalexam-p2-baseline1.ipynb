```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
train = pd.read_csv("train.csv")
X_test = pd.read_csv("test.csv")
submit = pd.read_csv("sample_submit.csv",index_col = 0)
```

```python
X_train = train.drop(['18~20_ride'],axis=1)
Y_train = train['18~20_ride']
```

```python
X_train = X_train.drop(['id','year','month','day'],axis=1)
X_test = X_test.drop(['id','year','month','day'],axis=1)
```

```python
# 2.24117
from sklearn.ensemble import RandomForestRegressor

clf = RandomForestRegressor(random_state=1)
clf.fit(X_train, Y_train)
y_test_pred = clf.predict(X_test)
```

```python
submit['18~20_ride'] = y_test_pred
submit.to_csv("submit_RF.csv")
```
