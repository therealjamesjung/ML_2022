```python
# 필요한 패키지 Import
import numpy as np 
import pandas as pd
```

```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
train = pd.read_csv("train.csv",encoding="latin-1")
test = pd.read_csv("test.csv",encoding="latin-1")
```

```python
x = train["Data"]
y = train["Class"]
test = test["Data"]
```

```python
from nltk.corpus import stopwords 
from nltk.stem.porter import PorterStemmer

def data_processing(text):
    pre_words = re.sub('[^A-Za-z]', ' ', text).lower()
    tokenized_words, stop = word_tokenize(pre_words), stopwords.words('english')
    tokenized_words_remove = [word for word in tokenized_words if word not in stop]
    stemmer = PorterStemmer()
    tokenized_words_remove = [stemmer.stem(word) for word in tokenized_words_remove]
    return (' '.join(tokenized_words_remove))
```

```python
x = [data_processing(i) for i in x]
test = [data_processing(i) for i in test]
```

```python
x
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_features=100)
x, test = np.array(x, dtype='U'), np.array(test, dtype='U')
x_features = vectorizer.fit_transform(x)
test_features = vectorizer.transform(test)
```

```python
y[y == "ham"] = 0
y[y == "spam"] = 1
y = y.astype("uint8")
```

```python
x_features
```

```python
from sklearn.svm import SVC

svc=SVC()
svc.fit(x_features, y)
res = svc.predict(test_features)
submit = pd.read_csv('sample_submit.csv')
submit['Class'] = res
submit.to_csv('submit.csv' ,index=False)
```
